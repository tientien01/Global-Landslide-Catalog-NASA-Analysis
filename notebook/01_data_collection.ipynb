# Thu thập dữ liệu

## Dữ liệu thuộc chủ đề nào?

- Bộ dữ liệu tập trung vào lĩnh vực **Khoa học Trái Đất** và **Quản lý Thiên tai**
- Cụ thể thì bộ dữ liệu này ghi lại các sự kiện sạt lở đất xảy ra trên toàn thế giới. Mỗi bản ghi đại diện cho một sự kiện sạt lở, bao gồm vị trí địa lý, thời gian xảy ra, nguyên nhân kích hoạt (chủ yếu là mưa) và mức độ thiệt hại về người (số người chết, bị thương).
- Bối cảnh thực tế:
    + Dữ liệu này đại diện cho bối cảnh thực tế về tần suất và sự phân bố của các thảm họa sạt lở đất trên thế giới. Nó phản ánh mối liên hệ giữa các điều kiện tự nhiên (như lượng mưa cực đoan) và các thảm họa địa chất ảnh hưởng đến con người.
    + Sạt lở đất là một trong các loại thiên tai phổ biến nhất, gây thiệt hại nghiêm trọng về người, cơ sở hạ tầng và kinh tế.
    + Bộ dữ liệu phản ánh rủi ro thiên tai, tác động của lượng mưa, địa hình, và tác động của biến đổi khí hậu lên các khu vực dễ bị tổn thương.
    + Đây là dữ liệu trợ giúp cho phân tích nguy cơ, quản lý thiên tai, và nghiên cứu địa chất – địa lý.

## Nguồn dữ liệu là gì?

- Dataset này được lấy từ:
    + Nền tảng: Kaggle
    + Tên dataset: Global Landslide Catalog
    + URL: https://www.kaggle.com/datasets/sathyanarayanrao89/global-landslide-catalog
- Tác giả gốc và tổ chức:
    + Tạo ra bộ dữ liệu này là NASA (National Aeronautics and Space Administration)
    + Cụ thể: NASA Goddard Space Flight Center – Landslide Hazard Assessment team
    + Đây là phiên bản sao chép từ dataset NASA GLC
- Ngày xuất bản/thu thập: Dựa trên mô tả bộ dữ liệu gốc của NASA, dữ liệu này thường bao gồm các sự kiện từ năm 2007 đến 2016

## Dữ liệu này có được cấp phép sử dụng không?

- Tập dữ liệu này có giấy phép: **"CC0: Public Domain"**
- Hoàn toàn được phép sử dụng cho mục đích giáo dục
- Hạn chế sử dụng hoặc yêu cầu ghi nhận: 
    + Không có hạn chế nghiêm ngặt về bản quyền
    + Về mặt đạo đức học thuật, cần trích dẫn nguồn gốc từ **“NASA Global Landslide Catalog, NASA Goddard Space Flight Center (GLC)”**


## Dữ liệu này được thu thập như thế nào?

- Phương pháp thu thập chính là tổng hợp thủ công và bán tự động từ các phương tiện truyền thông: Các nhà nghiên cứu tại NASA đã thu thập thông tin từ các báo cáo tin tức trực tuyến, cơ sở dữ liệu thảm họa, và các nguồn thông tin đại chúng khác để xác định các vụ sạt lở.
- Đối tượng là các sự kiện sạt lở đất liên quan đến mưa có tác động đáng kể (được báo chí đưa tin) trên toàn cầu.
- Khoảng thời gian thu thập dữ liệu: Từ năm 2007 đến 2016
- Bất kỳ hạn chế hoặc sai lệch trong quá trình thu thập:
    + Bias về ngôn ngữ: Vì dữ liệu dựa nhiều vào báo chí tiếng Anh, nên các vụ sạt lở ở các quốc gia không nói tiếng Anh hoặc vùng sâu vùng xa có thể bị thiếu sót.
    + Những nơi có truyền thông mạnh (Mỹ, châu Âu, Trung Quốc) có nhiều sự kiện được ghi nhận hơn. Các nước nghèo có thể có sự kiện thực nhưng không có báo cáo.
    + Độ chính xác vị trí: Một số điểm dữ liệu có thể chỉ mang tính ước lượng (ví dụ: bán kính 10km, 25km) chứ không phải tọa độ GPS chính xác từng mét.
    + Bias về quy mô: Các vụ sạt lở nhỏ không gây thiệt hại lớn thường không được lên báo, do đó có thể không xuất hiện trong dữ liệu này. Vùng núi, rừng sâu, xa dân cư thường thiếu dữ liệu.

## Tại sao chọn tập dữ liệu này?

- Nhóm quan tâm đến việc biến đổi khí hậu đang làm gia tăng các hiện tượng thời tiết cực đoan như mưa lớn, dẫn đến sạt lở đất. Việc hiểu rõ về sạt lở đất giúp nâng cao nhận thức về an toàn. Đây cũng là một hiện tượng thiên tai nghiêm trọng, ảnh hưởng đến tính mạng con người. Dataset lớn, thực tế nên có thể đưa ra được các phân tích phong phú hơn, xây dựng mô hình tốt hơn.
- Dữ liệu này có thể cung cấp những câu hỏi hoặc insights tiềm năng như là:
    + Các vụ sạt lở thường xảy ra ở những quốc gia nào?
    + Thời điểm nào trong năm dễ xảy ra sạt lở hơn?
    + Theo thời gian các vụ sạt lở đất có tăng không?
    + Những vụ sạt lở nào gây ảnh hưởng nghiêm trong đến tính mạng con người?
